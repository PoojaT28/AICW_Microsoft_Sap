```python
#face detection

#import opencv library for image processing

import cv2

#import cv2_imshow to display images in google colab
from google.colab.patches import cv2_imshow

#-------------------
#step 1:Load image
#------------------

#Path of teimages file(output this path as needed)

img_path='amitab.jfif'

#read the image from the given path
#OpenCv loads image in BGR Format
img=cv2.imread(img_path)

#check if the image was loaded successfully
#if the path is wrong and file not found, img will be none

if img is None:
  print(f"Error: Unable to loading image from {img_path}")
  exit()

#-------------------
#step 2:Load Haar Cascade for face detection
#------------------

#load the pretrained Haar cascade xml file for frontal face detection
#cv2.data.Haarcascades gives the default path where OpenCv stores this files

face_cascade= cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

#-------------------
#step 13:Convert image to grayscale
#------------------
#Haar Cascade works on grayscale images, not color images

gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#-------------------
#step 4:Detect faces in the images
#------------------
#detectMultiscale() detect objects(faces) of different sizes

faces=face_cascade.detectMultiScale(
    gray_img,       #Grayscale image
    scaleFactor=1.1,      #How much the image size is reduced at each scale
    minNeighbors=5,       #Higher value = fewerfalse positive
    minSize=(30,30)       #Minimum size of the object to be detected
)

#"faces" contains a list of rectangles:
#(x,y)--> top left corner
#w---> width of the rectangle
#h--> height of the rectangle

#-------------------
#step 5:Draw rectangles around faces
#------------------
#Loop through all detected faces

for(x,y,w,h) in faces:
  #Draw a rectangle around each detected face
  #(255,0,0) --> blue color in BGR format
  #Thickness of the rectangle
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

#-------------------
#step 6:Display the result (colab safe)
#------------------

cv2_imshow(img)


#-------------------
#step 7:save the output image(Optional)
#------------------

cv2.imwrite('detected_faces.jpg',img)


```
